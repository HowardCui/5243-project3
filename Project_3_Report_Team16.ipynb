## Introduction & Research Question

<div style="line-height: 1.8;">
User interface design plays an important role in whether users can successfully complete a multi-step task in a web application. This is especially true for first-time or occasional users, who may not yet understand the workflow or know what action to take next. For non-expert occasional users, a goals-guided interface was found to be more efficient and less error-prone than a conventional direct-manipulation interface, and users also showed a stronger preference for the guided version (Carrillo & Falgueras, 2020). This project evaluates whether guided navigation improves user progression and task completion in a Shiny-based data analysis app. This is the main research question of the study. Two versions of the app were developed with the same core functionality but different interface designs. Version A presents information in a flatter, more exploratory layout without step-by-step guidance, while Version B includes a more structured task flow, directional cues, and highlighted action buttons with helper text to support users as they move through the app. In both versions, users complete the same general workflow: loading a dataset, reviewing the data, navigating to the exploratory data analysis section, and generating a visualization. The purpose of this experiment is to examine how these two interface designs affect users’ ability to move through key stages of the task. More specifically, the study focuses on whether guided navigation helps users complete important steps in the workflow, successfully uploading data and generating a plot, compared with a less guided interface. 
</div>


## Experimental Design & Methodology

### Study Design


<div style="line-height: 1.8;">

This study was conducted as a randomized A/B test to evaluate whether guided navigation affects user progression through a Shiny-based data analysis app. In experimental-design terms, the study can also be described as a single-factor comparative experiment, because only one designed factor was intentionally manipulated: **interface version**. The purpose of the experiment was to determine whether a more structured interface could improve users’ ability to move through a short multi-step analysis task.

</div>


### Experimental Factor and Interface Conditions


<div style="line-height: 1.8;">

The experimental factor was **interface version**, with two levels: **Version A** and **Version B**. Both versions were implemented as Shiny for Python web applications and provided the same core analytical functions, including dataset loading, data review, and visualization generation. Therefore, the principal designed difference between the two conditions was the level of navigation guidance rather than the availability of different tools.

**Version A** served as the less-guided condition. It used a flatter and more exploratory interface layout in which users selected a data source and then navigated to the next step on their own. The layout emphasized the core controls and outputs but did not provide an explicit step-by-step workflow. As a result, users were expected to infer the task sequence more independently.

**Version B** served as the guided condition. It presented the same analytical workflow with more structured navigation support, including a visible user guide, clearer step order, and directional cues that highlighted the intended sequence of actions. Compared with Version A, this version was designed to reduce ambiguity about what users should do next and to support task completion through interface guidance rather than through additional analytical functionality.

### Randomization and Experimental Unit

<div style="line-height: 1.8;">

Users entered the experiment through a redirect entry point that randomly assigned each session to either Version A or Version B. The assigned interface condition was maintained during the active session so that users experienced only one version during a given session. The experimental unit for analysis was one anonymous app session, and each valid session contributed one observation to the final comparison between the two interface conditions.

</div>

### Task Procedure


<div style="line-height: 1.8;">

All users were given the same task: **load a dataset and create at least one visualization using the app**. Before formal data collection began, a small internal pilot test was conducted to confirm that the redirect logic, interface rendering, and event logging were functioning correctly. After the pilot phase, the same task procedure was used for all formal sessions.

</div>


### Operational Definition of Task Completion


<div style="line-height: 1.8;">

Task completion was defined behaviorally rather than through self-report. A session was considered successful if the user progressed through the workflow by loading a dataset and generating at least one valid chart in the exploratory data analysis section. This definition allowed task completion to be measured consistently across sessions using logged interaction events.

</div>


### Instrumentation and Event Logging


<div style="line-height: 1.8;">

User interactions were recorded through custom event logging integrated into the app and collected through Google Analytics 4 for validation and export. The logged events captured major workflow actions, including session start, upload success, visualization creation, and task completion. These event records were later transformed into a session-level analysis dataset so that stage-based user progression could be compared between the two interface conditions.

</div>

### Analytical Focus


<div style="line-height: 1.8;">

The methodology of this experiment was designed to support a direct comparison between the two interface versions on stage-based workflow outcomes. Because the two versions shared the same analytical functions, observed differences in user progression could be interpreted primarily as the effect of interface guidance. The numerical results and formal statistical comparisons are presented in the following section.

</div>


## Data Collection

<div style="line-height: 1.8;">
Data were collected from real users recruited by the project team through convenience sampling, primarily from classmates and friends, in order to obtain enough observations for meaningful comparison between the two interface versions. Participation was informal and voluntary, and all sessions were recorded anonymously. No personally identifiable information was collected. When users entered the app, they were randomly assigned to either Version A or Version B through the redirect entry point, ensuring that each participant experienced only one interface condition during a session. During the data collection period, Google Analytics recorded 163 active users. The collected data included anonymous group assignment, session-level identifiers, and event-based measures of user progression through the app workflow. In particular, the data captured whether users reached the upload stage and whether they successfully created a visualization, which were later used in the stage-based statistical analysis.
</div>

## Statistical Analysis & Results

Although Google Analytics recorded 163 active users during the broader data collection period, the final analysis was based on **160 users at the session-start stage**, including **75 users in Group A** and **85 users in Group B**.

To compare user conversion rates between Version A and Version B, we conducted **two-proportion z-tests** for two stages of the workflow:

1. **Stage 1:** start $\rightarrow$ upload  
2. **Stage 2:** upload $\rightarrow$ create plot 
 

Table 1. Stage counts by group across workflow stages.

|---------------|--------:|--------:|------:|
| Start         | 75      | 85      | 160   |
| Upload        | 51      | 42      | 93    |
| Visualization | 32      | 36      | 68    |


### Hypotheses

For each stage, let $p_A$ and $p_B$ denote the true completion proportions for Groups A and B.

The null and alternative hypotheses were:

$$
H_0: p_A = p_B
$$

$$
H_A: p_A \ne p_B
$$

A significance level of

$$
\alpha = 0.05
$$

was used for both tests.

### Analysis


<div style="line-height: 1.8;">

For both stages, two-proportion z-tests were conducted to compare the conversion rates between groups A and B. The null hypothesis assumed that there was no difference in proportions between the two groups ($H_0: p_A = p_B$), while the alternative hypothesis assumed a difference ($H_A: p_A \ne p_B$). A significance level of $\alpha = 0.05$ was used. As shown in **Table 1**, the final stage counts were 75 vs. 85 users at the start stage, 51 vs. 42 users at the upload stage, and 32 vs. 36 users at the visualization stage for Groups A and B, respectively. The corresponding statistical test results are summarized in **Table 2**.

For the first stage (**start $\rightarrow$ upload**), the test resulted in a p-value of **0.0174**. Since $p < 0.05$, we reject the null hypothesis and conclude that there is a statistically significant difference between the two groups. Specifically, users in Group A had a higher probability of proceeding to the upload step compared to Group B.

For the second stage (**upload $\rightarrow$ create plot**), the p-value was **0.0129**. Again, since $p < 0.05$, we reject the null hypothesis. In this case, users in Group B were significantly more likely to complete the task after uploading, indicating that guided navigation improved later-stage completion.

In addition to hypothesis testing, effect sizes and confidence intervals were calculated to assess the magnitude and reliability of the observed differences. At the first stage (**start $\rightarrow$ upload**), the difference in completion rates between Group A and Group B was **0.1859** (**18.59 percentage points**), with a 95% confidence interval of **(0.0361, 0.3357)**. Since the interval does not include zero and is entirely positive, this indicates a statistically and practically significant advantage for Group A. In other words, users in the unguided interface were substantially more likely to proceed to the upload step.

At the second stage (**upload $\rightarrow$ create plot**), the effect size was **-0.2297** (**-22.97 percentage points**), with a 95% confidence interval of **(-0.3994, -0.0600)**. The interval is entirely negative, indicating that Group B significantly outperformed Group A at this stage. This suggests that guided navigation improved task completion once users had progressed beyond the initial step.

The results reveal a clear trade-off between the two interface designs. While the unguided interface performed better in the initial stage by allowing users to proceed more easily, the guided interface improved completion in the later stage by providing structured support. This suggests that guided navigation may introduce additional cognitive or interaction burden at the beginning, but becomes beneficial once users are engaged in the workflow. Overall, the findings indicate that the effectiveness of guided navigation depends on the stage of user interaction rather than being universally beneficial.

## Interpretation & Conclusion

### Interpretation of Findings

The results show that the two interface versions performed differently at different stages of the workflow. In the first stage, from **start to upload**, Group A had a significantly higher conversion rate than Group B. This suggests that the less-guided interface may have made it easier for users to move quickly into the app and begin interacting with the data upload step. One possible explanation is that the flatter layout created less interruption at the beginning of the workflow and allowed users to proceed more directly.

In the second stage, from **upload to create plot**, the pattern was reversed. Group B had a significantly higher conversion rate than Group A. This suggests that once users had already entered the workflow, the guided interface provided more useful support for completing the later task. The clearer step order, directional cues, and helper text may have reduced uncertainty and helped users continue from data upload to successful visualization creation.

Taken together, the results suggest that the effect of guided navigation is stage-dependent rather than uniform across the whole workflow. The unguided version appeared to support earlier progression, while the guided version appeared to support later-stage completion. Therefore, the results do not suggest that one interface was better in every respect. Instead, they show that different design choices may be more effective at different points of the user journey.

### Practical Implications

From a practical design perspective, these findings suggest that a fully guided or fully unguided interface may not be the best solution for all parts of a task-based data analysis app. A more balanced design may be more effective. For example, the app could keep a simple and low-friction entry experience at the beginning, while adding more explicit guidance after users begin the task. In other words, a hybrid design may combine the strengths of both versions.

This conclusion is relevant to the broader goal of improving user experience in interactive analytical tools. For first-time or occasional users, the challenge is not only understanding what the app can do, but also knowing what action to take next. The current findings suggest that guidance is especially helpful after the user has already committed to the workflow. This provides useful evidence for future interface refinement.

### Conclusion

In conclusion, this project found meaningful differences between Version A and Version B in stage-based user progression. Version A performed better in the transition from session start to upload, while Version B performed better in the transition from upload to successful plot creation. These results suggest that interface guidance influences user behavior, but its effect depends on the stage of interaction. Overall, the study provides initial evidence that guided navigation can improve later-stage task completion, although it may also introduce some friction at the beginning of the workflow. Future design work should therefore focus on combining ease of entry with stronger support during task completion.

## Challenges & Limitations

### Sample and Generalizability

A first limitation of this study is the sampling approach. Participants were recruited mainly through classmates, friends, and other available users. This convenience sample was appropriate for a course project, but it limits the generalizability of the findings. The users in this study may not fully represent the broader population of potential users of a data analysis app, especially users with different technical backgrounds, task goals, or levels of experience.

### Data Scope and Measurement Constraints

A second limitation is related to the type of exported data used in the final analysis. The reported results were based mainly on funnel exploration data, which allowed comparison of conversion rates at key workflow stages. This was suitable for the stage-based A/B analysis reported in the paper. However, this type of data is more limited than a full session-level export. As a result, not all originally planned outcomes could be examined in the final report, especially more detailed analyses of time to completion or richer user-level behavioral patterns.

### Design and Causal Specificity

Although the study used randomized assignment between Version A and Version B, the two versions differed in several interface features at the same time, including workflow structure, directional cues, and helper text. Therefore, the experiment can identify the effect of the overall interface condition, but it cannot isolate which single design element caused the observed differences. In future work, a more targeted A/B design could test one interface component at a time, such as guidance text alone or navigation layout alone.

### Analytical Scope

The current analysis focused on stage-based conversion outcomes, which are meaningful and relevant to the project goal. However, these results should not be interpreted as a complete evaluation of all possible aspects of usability. The final analysis does not fully capture long-term engagement, repeated sessions from the same person, or deeper differences in task quality beyond successful stage progression. Future work with a larger and cleaner session-level dataset could support more complete analyses, including end-to-end completion, time-based outcomes, and stronger robustness checks.

### Device Compatibility and Interface Context

Another limitation is that the deployed Shiny app was mainly designed and reviewed in a desktop-based setting. The interface was not specifically optimized for mobile users or for different operating systems such as iOS and Android. As a result, some layout, navigation, or interaction differences may have been experienced differently across devices. This may have influenced user behavior in the early stage of the workflow and could partly explain why Group A showed a higher conversion rate than Group B in the start-to-upload stage. Because device type was not formally controlled or analyzed in the current study, this possible explanation cannot be tested directly in the present report.

Despite these limitations, the project still provides useful evidence about how interface structure shapes user progression in a multi-step analytical app. The study successfully implemented a real A/B testing framework, collected original behavioral data, and identified meaningful differences between the two interface conditions. For a course-based experimental project, this provides a strong foundation for future improvement and further testing.
