# Introduction & Research Question

<div style="line-height: 1.8;">
User interface design plays an important role in whether users can successfully complete a multi-step task in a web application. This is especially true for first-time or occasional users, who may not yet understand the workflow or know what action to take next. For non-expert occasional users, a goals-guided interface was found to be more efficient and less error-prone than a conventional direct-manipulation interface, and users also showed a stronger preference for the guided version (Carrillo & Falgueras, 2020). This project evaluates whether guided navigation improves user progression and task completion in a Shiny-based data analysis app. Two versions of the app were developed with the same core functionality but different interface designs. Version A presents information in a flatter, more exploratory layout without step-by-step guidance, while Version B includes a more structured task flow, directional cues, and highlighted action buttons with helper text to support users as they move through the app. In both versions, users complete the same general workflow: loading a dataset, reviewing the data, navigating to the exploratory data analysis section, and generating a visualization. The purpose of this experiment is to examine how these two interface designs affect users’ ability to move through key stages of the task. More specifically, the study focuses on whether guided navigation helps users complete important steps in the workflow, successfully uploading data and generating a plot, compared with a less guided interface. 
</div>

# Data Collection

<div style="line-height: 1.8;">
Data were collected from real users recruited by the project team through convenience sampling, primarily from classmates and friends, in order to obtain enough observations for meaningful comparison between the two interface versions. Participation was informal and voluntary, and all sessions were recorded anonymously. No personally identifiable information was collected. When users entered the app, they were randomly assigned to either Version A or Version B through the redirect entry point, ensuring that each participant experienced only one interface condition during a session. During the data collection period, Google Analytics recorded 163 active users. The collected data included anonymous group assignment, session-level identifiers, and event-based measures of user progression through the app workflow. In particular, the data captured whether users reached the upload stage and whether they successfully created a visualization, which were later used in the stage-based statistical analysis.
</div>
